```python
# Load packages

import numpy as np
import scipy as sp
from scipy.stats import moment,describe
import matplotlib.pyplot as plt
import skimage
from skimage import data,color,feature,filters
from skimage.segmentation import slic,mark_boundaries

np.set_printoptions(precision=4)
```

```python
# Segmentation and masking functions

# segments img into nseg segments using k-means clustering
def segment(img,nseg,comp=1,sig=0):
    seg = slic(img,n_segments=nseg,compactness=comp,sigma=sig,convert2lab=True,enforce_connectivity=False) 
    
    return seg

# marks the boundaries of a segmented image
def mark(img,seg):
    return mark_boundaries(img,seg)

# creates boolean masks for each segment
def getMasks(img,seg):
    masks = []

    for (i,segVal) in enumerate(np.unique(seg)):
        mask = np.zeros(img.shape,dtype=bool)
        mask[seg == segVal] = True
        masks.append(mask)

    return masks

# creates images for each segment
def maskImgs(img,masks):
    masked = []

    for mask in masks:
        temp = np.copy(img)
        temp[np.logical_not(mask)] = 0
        masked.append(temp)

    return masked

# converts a row,col,3 rgb mask to a row,col,1 gray mask
def grayMask(rgbmask):
    return rgbmask[:,:,0]
```

```python
# HSV statistics functions

# calculates the 1st 4 moments of a component of an hsv image
def getStats(hsv,mask,param=0):
    par = hsv[:,:,param]
    newmask = mask[:,:,0]
    count = par[newmask]
    stat = describe(count)
    
    return stat[0],stat[2],stat[3],stat[4],stat[5]

# calculates the 1st 4 moments of a component of all of the hsv images
def getAllStats(hsv,masks,param=0):
    stats = []
    
    for mask in masks:
        stats.append(getStats(hsv,mask))
    
    return stats


# computes hue frequency histogram for hsv image
# param = 0 for hue, param=1 for saturation
def hsvHist(hsv,mask,param=0,nbins=30):
    par = hsv[:,:,param]
    newmask = mask[:,:,0]
    count = par[newmask]
    return np.histogram(count,bins=nbins,density=True)

# computes histograms for list of hsv images
# param = 0 for hue, param=1 for saturation
def hsvHistAll(hsv,masks,param=0,nbins=32):
    hists = []
    for mask in masks:
        hists.append(hsvHist(hsv,mask,param,nbins))

    return hists

# display 2 images side by side
def display(img1,img2):
    fig,ax = plt.subplots(nrows=1,ncols=3)
    ax[0].imshow(img1)
    ax[1].imshow(img2)
    ax[0].set_xticks([]); ax[0].set_yticks([]); ax[1].set_xticks([]); ax[1].set_yticks([]);
    
    return

# display all images in a list
def displayAll(imgs):

    for im in imgs:
        fig,ax = plt.subplots()
        ax.imshow(im)
        ax.set_xticks([]); ax.set_yticks([]);
    return

def getShape(bins):
    center = (bins[:-1] + bins[1:]) / 2
    width = 1.0*(bins[1] - bins[0])
    return center,width

def plotImgHist(img,hist,center,wid):
    fig,ax = plt.subplots(nrows=1,ncols=2)
    ax[0].imshow(img); ax[0].set_xticks([]); ax[0].set_yticks([]);
    ax[1].bar(center,hist,width=wid)
    ax[1].set_xlim([0,1]); ax[1].set_ylim([0,1.05*np.max(hist)]);

    return

def displayMaskAndHist(masked,hsvHists):

    for i in range(len(masked)):
        hist,bins = hsvHists[i]
        center,width = getShape(bins)
        plotImgHist(masked[i],hist,center,width)
    return
```

```python
# Load data

# import sample images
astro = data.astronaut() 
ihc = data.immunohistochemistry()
rock = data.rocket()
cat = data.chelsea()

# import external images
name = "rust_insulators"
#name = "insulators"
#name = "tree"
#name = "rust"
#name = "rust2"

form = ".jpg"
fname = "../data/"+name+form
file = data.imread(fname)
```

```python
# Select image and parameters

img = file # select which image

if img.ndim == 2: # convert grayscale to rgb
    img = color.gray2rgb(img)
hsv = color.rgb2hsv(img) # get hsv format of img
gray = skimage.util.img_as_ubyte(color.rgb2gray(img)) # get grayscale format of img

nbins = 64 # number of histogram bins
```

```python
# Image preprocessing
mask = np.ones(hsv.shape,dtype=bool)
hist,bins = hsvHist(hsv,mask,param=0,nbins=nbins)
center,width = getShape(bins)

plotImgHist(img,hist,center,width)
plt.show()
```

```python
# Color analysis

# run color analysis of image
def colorAnalysis(img,k,plot=True):
    seg = segment(img,k,comp=0.1,sig=1) # segment image
    marked = mark(img,seg) # mark boundaries
    masks = getMasks(img,seg) # get bool masks for each segment
    masked = maskImgs(img,masks) # get masked images
    hsvHists = hsvHistAll(hsv,masks,param=0,nbins=nbins) # compute hue histograms
    
    if plot:
        display(marked,seg) # display original image with segment boundaries marked
        displayMaskAndHist(masked,hsvHists) # display histograms with images
    
    return masks,hsvHists

k = 4
%time masks,hsvHists = colorAnalysis(img,k,plot=True)

stats = getAllStats(hsv,masks)
for line in stats: print(np.asarray(line))

plt.show()
```

```python
# Filters

mask=None
#mask = graymask(masks[1])

edgel = filters.laplace(gray,mask=mask)
edgep = filters.prewitt(gray,mask)
edger = filters.roberts(gray,mask)
edges = filters.scharr(gray,mask)
edgeso = filters.sobel(gray,mask)


display(img,edgel)
display(edgep,edger)
display(edges,edgeso)
plt.show()
```

```python
# Features

mask=None
#mask = masks[0]

can = feature.canny(gray,mask=mask)


display(img,can)
plt.show()
```
